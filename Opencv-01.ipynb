## ðŸ“„ `cv2.imread()` Parameters

| Parameter  | Type  | Description                                                                          |
| ---------- | ----- | ------------------------------------------------------------------------------------ |
| `filename` | `str` | Path to the image file to be loaded.                                                 |
| `flags`    | `int` | (Optional) Specifies the way image should be read. Default is `cv.IMREAD_COLOR` (1). |


```python
import cv2 as cv 
import numpy as np
img = cv.imread('flower-sample.jpg')


cv.imshow('image', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

### ðŸ“Œ Common `flags` Values for `cv2.imread()`

| Flag Name                   | Value | Description                                                                                 |
| --------------------------- | ----- | ------------------------------------------------------------------------------------------- |
| `cv.IMREAD_COLOR`           | `1`   | Loads a color image. Any transparency of image will be neglected. This is the default flag. |
| `cv.IMREAD_GRAYSCALE`       | `0`   | Loads image in grayscale mode.                                                              |
| `cv.IMREAD_UNCHANGED`       | `-1`  | Loads image as such including alpha channel (transparency), if present.                     |
| `cv.IMREAD_ANYDEPTH`        | `2`   | Loads image in any bit depth.                                                               |
| `cv.IMREAD_ANYCOLOR`        | `4`   | Loads image in any color format.                                                            |
| `cv.IMREAD_REDUCED_COLOR_2` | `17`  | Loads the image reduced by 1/2.                                                             |
| `cv.IMREAD_REDUCED_COLOR_4` | `33`  | Loads the image reduced by 1/4.                                                             |
| `cv.IMREAD_REDUCED_COLOR_8` | `65`  | Loads the image reduced by 1/8.                                                             |

## ðŸ“ Resizing Images in OpenCV


### ðŸ“„ Parameters of `cv2.resize()`

| Parameter       | Type               | Description                                                                                     |
| --------------- | ------------------ | ----------------------------------------------------------------------------------------------- |
| `src`           | `ndarray`          | Source image (the image to be resized).                                                         |
| `dsize`         | `tuple`            | Desired size as `(width, height)`. If set to `(0, 0)`, it must be calculated via `fx` and `fy`. |
| `fx`            | `float` (optional) | Scale factor along the horizontal axis (width). Used when `dsize` is not specified.             |
| `fy`            | `float` (optional) | Scale factor along the vertical axis (height). Used when `dsize` is not specified.              |
| `interpolation` | `int` (optional)   | Interpolation method used to resize the image. Default is `cv2.INTER_LINEAR`.                   |

---

### ðŸ”€ Interpolation Methods

| Interpolation Flag   | Description                                                                 |
| -------------------- | --------------------------------------------------------------------------- |
| `cv2.INTER_NEAREST`  | Nearest neighbor interpolation (fastest, lower quality).                    |
| `cv2.INTER_LINEAR`   | Bilinear interpolation (default; good for upscaling).                       |
| `cv2.INTER_AREA`     | Resampling using pixel area relation (good for downscaling).                |
| `cv2.INTER_CUBIC`    | Bicubic interpolation over 4x4 pixel neighborhood (slower, better quality). |
| `cv2.INTER_LANCZOS4` | Lanczos interpolation over 8x8 pixel neighborhood (high-quality).           |

---

### âš ï¸ Notes

* Either `dsize` **or** `fx`/`fy` must be provided.
* If both are provided, `dsize` takes priority and `fx`, `fy` are ignored.



```python
img_resize = cv.resize(rand_img,(400,400))
cv.imshow('image', img_resize)
cv.waitKey(0)
cv.destroyAllWindows()
```
